1. Extraer la información del archivo.

Se importa la libreria pandas y se coloca el alias pd, ademas se lee el archivo "country_vaccinations.csv" y se imprime en pantalla

```python
import pandas as pd

#from google.colab import drive
#drive.mount('/content/drive')

vacunas = pd.read_csv("country_vaccinations.csv")
vacunas
```

2. Estructuras y tipos de datos

Se muestra la estructura y tipos de datos de cada columna para identificar qué operaciones se pueden realizar con cada una de ellas, ademas validadmos que las columnas de fecha sean del tipo datetime64, con la funcion to_datetime.

```python
vacunas['date'] = pd.to_datetime(vacunas['date'])
vacunas.info()
```

3. Determinar la cantidad de vacunas aplicadas de cada compañía

 (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
cantidad_vacunasporcompania = (vacunas.groupby("vaccines")["daily_vaccinations"].sum().reset_index(name="cantidad_vacunas_por_compania"))
cantidad_vacunasporcompania
```

4. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
total_vacunas = vacunas['daily_vaccinations'].sum()
print(f'Total de vacunas aplicadas en el mundo:  { total_vacunas}')
```

5 Calcular el promedio de vacunas aplicadas por país.

```python
promedio_vacunas = (vacunas.groupby('country')['daily_vaccinations'].mean().reset_index(name='Promedio_diario'))
promedio_vacunas
```

6. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
vacunas_diaespecifico = vacunas.query('date==@pd.Timestamp("2021-01-29")')
totalvacunas_diaespecifico = vacunas_diaespecifico['daily_vaccinations'].sum()
print(f'Total de vacunas aplicadas el dia 29/01/21: { totalvacunas_diaespecifico }');
```

7. Creacion de dataframe conDiferencias

Creamos un dataframe nuevo denominado conDiferencias que contiene los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias = vacunas.copy()
conDiferencias['diferencia'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias
```

8. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
primeraFecha = vacunas['date'].min()
ultimaFecha = vacunas['date'].max()
periodo = ultimaFecha - primeraFecha
print(f"El periodo de tiempo entre la fecha mas reciente y las mas antigua es: { periodo }")
```

9. Creacion de un dataframe nuevo denominado conCantidad

este contiene los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (se usa la columna vaccines separando los carácter por la "," )

```python
conCantidad = vacunas.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split(",").str.len()
conCantidad
```

10. Generar un dataframe denominado antes20

con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = vacunas.query('date<@pd.Timestamp("2020-12-20")').copy()
antes20
```

11. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = vacunas[vacunas['vaccines'].str.contains('Pfizer')].copy()
pfizer
```

12. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente

```python

with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
